<p style="color:SlateBlue;">
In the empty Code cell below, provide the necessary coding to create a plot based on the linear equation y = 5x - 6. Also, give your plot a suitable title by change the labels for the x and y axes to be Height (cm) and Mass (kg), the title of the plot to "Object Height (cm) vs Mass(kg)", espectively.</p>

--------------------------

```python
import matplotlib.pyplot as plt
import numpy as np

## Now we create our figure and axes for the plot we will make.
fig, ax = plt.subplots(figsize=(10,6))

## Plot a straight according to linear equation y = 5x - 6
x = np.linspace(-2,12)
y = 2*x - 6

## Label the axis and title
ax.set_xlabel("Height (cm)")
ax.set_ylabel("Mass (kg)")
ax.set_title("Object Height (cm) vs Mass (kg)")

## Plot according to the data generated, connecting the points with lines
ax.plot(x, y)

## We call plt.show() to show the chart (optional, but good to have)
plt.show()
```

<p style="color:SlateBlue;">In the empty Code cell below, provide the necessary codes for the following requirements:</p>
<ol style="color:DodgerBlue;">
    <li>Plot the linear function y = 18x - 5 and the quadratic function y = -7x^2 + 3 on the same figure.</li>
    <li>Specify both the range limitation for the x-axis and y-axis (i.e. allow spaces on the left, right, top, and bottom) to enhace the readability.</li>
    <li>Specify the ticks to increase the # of intervals.</li>
    <li>Label the title of the plot as "Sensitibity Analysis".</li>
    <li>Label the x-axis as "Costs of Advertisement".</li>
    <li>Label the x-axis as "Marginal Profits".</li>
    <li>Save the plot to the file "w13pe2.pdf" for future reference. (check to ensure the file not exists before save your plot to there).</li>
    </ol>


----------------------------------------

```python
import matplotlib.pyplot as plt
import numpy as np
import os

## Now we create our figure and axes for the plot we will make.
fig, ax = plt.subplots(figsize=(10,6))

## Plot a straight according to linear equation y = 18x - 5
x1 = np.linspace(-4, 2)
y1 = 18 * x1 - 5

## Plot the quadratic function y = -7x^2 + 3
x2 = np.linspace(-4, 2)
y2 = -7 * x2**2 + 3

## Label the axis and title
ax.set_xlabel("Costs of Advertisement")
ax.set_ylabel("Marginal Profits")
ax.set_title("Sensitivity Analysis")

## Specify the limit on both x-axis and yaxis
ax.set_xlim(-4.5, 2.5)
ax.set_ylim(-120, 40)

## specify the ticks to allow sufficient number of intervals
ax.set_xticks(np.arange(-4, 2, 0.5))
ax.set_yticks(np.arange(-120, 40, 20))

## Plot according to the data generated, connecting the points with lines
ax.plot(x1, y1)
ax.plot(x2, y2)

## We call plt.show() to show the chart (optional, but good to have)
plt.show()

## Save the file 
if os.path.isfile("w13pe2.pdf"):
    os.remove("w13pe2.pdf")
plt.savefig("w13pe2.pdf")
```
