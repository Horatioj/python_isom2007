<h3 style="color:blue;"><b>Introduction to Seaborn</b></h3>
<ul style="color:SlateBlue;font-size:14px;">
    <li>MatPlotLib modules provides powerful plotting functionality, but the MatPlotLib API can be rather daunting (令人生畏的) for beginners, and the basic color schemes and plot styles are not the most visually appealing (視覺吸引力).</li>
    <li>As a result, other libraries (one example is the Seaborn modules) have been developed that build on the MatPlotLib legacy (遺留) and provide both improved visual aesthetics (美學) as well as new plot functions.</li>
    <li>Seaborn, actually not difficult to start using, the current lecture we will introduce a couple of new powerful plot styles along with statistical functionality that can make difficult plotting tasks simple.</li>
    <li>Seaborn also simplifies the task of removing plot clutter (混亂), and you also can choose from different color palettes (調色板) that have already been screened to maximize the visual impact of our new plots.</li>
    <li>Seaborn can be easily applied into our Python program by using statements below to improve the plot’s colors and styles:</li>
    <span style="color:white;">.......</span><span style="color:orange;">import seaborn as sns</span><br>
    <span style="color:white;">.......</span><span style="color:orange;">sns.set()</span><br><br>
    <li>The set method takes a number of parameters like style or font that you can also specify explicitly.</li>
    <li>Referred to the following link for the details on the using of all other modules of Seaborn:</li>
    <span style="color:white;">.......</span><span style="color:brown;">https://www.w3schools.com/python/numpy_random_seaborn.asp</span>
    </ul>

```python
## Recall of the plot that we have drew in the previous lectures
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,5))
x1 = np.linspace(-5, 15)
y1 = 8 + 5 * x1
x2 = x1
y2 = 12 - 6 * x2
ax.plot(x1, y1, x2, y2)

## Set labels of axis
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")

## Set limit of the range on plot
ax.set_xlim(-7, 18)
ax.set_ylim(-100, 100)

## Set the ticks on each axis
ax.set_xticks(np.arange(-6, 18, 2))
ax.set_yticks(np.arange(-80, 100, 20))
    
# Set the title
ax.set_title("The Original Plot!")

plt.show()
```

```python
# Now Seaborn specific modifications
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

## Seaborn specifications, specify the context and font size
sns.set_context("poster", font_scale=0.5)     
sns.set_style("white")
sns.set_style("ticks")

# Now we create our figure and axes for the plot we will make.
fig, ax = plt.subplots(figsize=(10,5))
x= np.linspace(-5, 15)
y1 = 8 + 5 * x
y2 = 12 - 6 * x

## Plot the graph with the x and y that specified
ax.plot(x, y1, label = 'Plot 1', ls = ':')
ax.plot(x, y2, label = 'Plot 2', ls = 'dashdot')

## Set labels of axis
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")

## Set limit of the range on plot
ax.set_xlim(-7, 18)
ax.set_ylim(-100, 100)

## Set the ticks on each axis
ax.set_xticks(np.arange(-6, 18, 2))
ax.set_yticks(np.arange(-80, 100, 20))

## Set the legend and specify its location
plt.legend(loc='upper left')

## Set the title
ax.set_title("Plot with Seaborn Specifications!")

## Seaborn specific modifications
sns.despine(offset=6, trim=True)

## Display the plot accordingly
plt.show()
```

```python
## Recall of the plot that we have drew in the previous lectures with Seaborn setting
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns            

## Define some color for the graph
g_colors = [sns.xkcd_rgb['red'], sns.xkcd_rgb['denim blue'], sns.xkcd_rgb['bluegrey'], 
            sns.xkcd_rgb['faded green'], sns.xkcd_rgb['green']]

## Seaborn specifications
sns.set_context("poster", font_scale=0.5)      # specify the context with font size
sns.set_style("white")
sns.set_style("ticks")

fig, ax = plt.subplots(figsize=(10,5))
x1 = np.linspace(-5, 15)
y1 = 8 + 5 * x1
x2 = x1
y2 = 12 - 6 * x2
ax.plot(x1, y1, label = 'Plot 1', ls=':', color=g_colors[0])           # specify the style & color
ax.plot(x2, y2, label = 'Plot 2', ls="dashdot", color=g_colors[-1])
        
## Set labels of axis
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")

## Set limit of the range on plot
ax.set_xlim(-7, 18)
ax.set_ylim(-100, 100)

## Set the ticks on each axis
ax.set_xticks(np.arange(-6, 18, 2))
ax.set_yticks(np.arange(-80, 100, 20))

plt.legend(loc='upper left')             # include the legend and specify its location
    
## Set the title
ax.set_title("Plot with Seaborn Specifications!")

## Seaborn specific modifications
sns.despine(offset=10, trim=True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import math

sns.set_context('poster', font_scale=0.5)
sns.set_style("white")
sns.set_style("ticks")

fig, ax = plt.subplots(figsize=(20, 8))

x = np.linspace(-12, 8)
y1 = 4*x**2 + math.log(x, 10)
y2 = -3*x**2 -13*x - math.log(x, 10)
ax.plot(x, y1, label = "Plot I", ls="--", color=sns.xkcd_rgb['brown'])
ax.plot(x, y2, label = "Plot II", ls="-.", color=sns.xkcd_rgb['orange'])

ax.set_title("Plot", fontsize = 24)
ax.set_xlabel("Height(cm)", fontsize = 20)
ax.set_ylabel("Mass(kg)", fontsize = 20)

ax.set_xlim(-14, 10)
ax.set_ylim(-200, 200)
ax.set_xticks(np.arange(-14, 8, 1))
ax.set_yticks(np.arange(-200, 201, 50))

plt.legend("upper right")
```
