```python
## Standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

## We do this to ignore several specific Pandas warnings
import warnings
warnings.filterwarnings("ignore")

## Use default white plot style
sns.set(style="white")

## Scatter plot to display the correlation between mid-term with exam scores
## Input the Data set, and define our two dimensions
tpDF = pd.read_excel("tips.xlsx")
tsSet = tpDF['total_bill']
tpSet = tpDF['tip']

## Create our figure and axes for the plot and plot the data
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(tsSet, tpSet)

# Set our axis labels and title
ax.set_xlabel("Total Bills ($)", fontsize=12)
ax.set_ylabel("Tips ($)", fontsize=12)
ax.set_title("Comparison of Total Bills with Tips!", fontsize=14)
ax.set_xlim(0, 600)
ax.set_ylim(0, 120)

sns.despine(offset=10, trim=True)
plt.show()
```

<p style="color:SlateBlue;font-size:14px;">In the empty Code cell below, write and execute code to input the data from tips.xlsx / tips.csv,  make a new scatter plot that contains the compares of total_bill and tip items on both male and female.</p>

--------------------

```python
# Set up the Notebook working environment
%matplotlib inline

## Standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

## We do this to ignore several specific Pandas warnings
import warnings
warnings.filterwarnings("ignore")

## Use default white plot style
sns.set(style="white")

## Scatter plot to display the correlation between mid-term with exam scores
## Input the Data set, and define our two dimensions
tpDF = pd.read_excel("tips.xlsx")
tsSetM = tpDF.loc[tpDF.sex=='Male', ['total_bill']]
tpSetM = tpDF.loc[tpDF.sex=='Male', ['tip']]
tsSetF = tpDF.loc[tpDF.sex=='Female', ['total_bill']]
tpSetF = tpDF.loc[tpDF.sex=='Female', ['tip']]

## Create our figure and axes for the plot and plot the data
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(tsSetM, tpSetM, color=sns.xkcd_rgb["pale red"], marker='p', label='Male')
ax.scatter(tsSetF, tpSetF, color=sns.xkcd_rgb["blue"], marker='d', label='Female')

# Set our axis labels and title
ax.set_xlabel("Total Bills ($)", fontsize=12)
ax.set_ylabel("Tips ($)", fontsize=12)
ax.set_title("Comparison of Total Bills with Tips!", fontsize=14)
ax.set_xlim(0, 600)
ax.set_ylim(0, 110)

ax.legend(loc='upper left', fontsize=14)

sns.despine(offset=10, trim=True)
plt.show()
```

```python
## Standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as sts

## We do this to ignore several specific Pandas warnings
import warnings
warnings.filterwarnings("ignore")

## Use default white plot style
sns.set(style="white")

## Scatter plot to display the correlation between mid-term with exam scores
## Input the Data set, and define our two dimensions
tpDF = pd.read_excel("tips.xlsx")
tsSet = tpDF['total_bill']
tpSet = tpDF['tip']

## Create our figure and axes for the plot and plot the data
fig, ax = plt.subplots(figsize=(10,8))
##ax.scatter(tsSet, tpSet)
ax = sns.regplot(x='total_bill', y='tip', data=tpDF, fit_reg=True, ci=95)  # instead use of scatter plot

# Set our axis labels and title
ax.set_xlabel("Total Bills ($)", fontsize=12)
ax.set_ylabel("Tips ($)", fontsize=12)
ax.set_title("Comparison of Total Bills with Tips!", fontsize=14)
ax.set_xlim(0, 600)
ax.set_ylim(0, 120)

sns.despine(offset=10, trim=True)
plt.show()

# Compute our slope, intercept, and r value
slope, intercept, r_value, p_value, slope_std_error = sts.linregress(tsSet, tpSet)

# Display results
print(f'\nBest fit line: y = {slope:4.2f} x + {intercept:5.3f}')
print(f'Pearsonr correlation = {r_value:5.3f}')
```
