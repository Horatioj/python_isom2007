<h3 style="color:blue;"><b>Rug Plots</b></h3>
<ul style="color:SlateBlue;font-size=14px;">
    <li>A rugplot displays a one-dimensional distribution of data. </li>
    <li>In Python, we can make rugplots by using the rugplot function provided by the Seaborn module, which takes:</li>
    <ul style="color:SlateBlue;font-size=14px;">
        <li>a column from a Pandas DataFrame, such as tips['total_bill'], and draws short vertical lines along the x-axis for every data point in the column. </li>
        <li>height (optional), which specified how long (in inches) the vertical line should extend, </li>
        <li>lw (optional) for the thickness of the vertical line,</li>
        <li>color (optional) that specifies a color for the vertical lines. </li>
    </ul>
    </ul>

----------------

```python
%matplotlib inline

# These two lines suppress warnings that sometimes
# occur when making visualizations
import warnings
warnings.filterwarnings('ignore')
```

```python
## Rugplot Plotting
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

## Input the data into pandas DataFram
tipDF = pd.read_excel("tips.xlsx")

## Define plot layout
fig, ax = plt.subplots(figsize=(10, 1.5))

## Make the rugplot
sns.rugplot(ax=ax, a=tipDF['total_bill'], height=0.25, lw=0.5, \
            color = sns.xkcd_rgb["pale red"]) ;

## Display the plot
plt.show()
```

```python
## Rugplot Plotting (use set_title to name the title)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

## Input the data into pandas DataFram
tipDF = pd.read_excel("tips.xlsx")

## Define plot layout
fig, ax = plt.subplots(figsize=(10, 1.5))

## Make the rugplot
sns.rugplot(ax = ax, a=tipDF['total_bill'], height=0.25, lw=0.5, \
            color = sns.xkcd_rgb["pale red"]) ;

## Configure the Rugplot outlook
ax.set_xlim(-50, 600)
ax.set_xlabel("Total Bills")
ax.set_title("RugPlot on Total Bills Distribution", ha='center', \
             fontsize=20, color=sns.xkcd_rgb["pale red"])

## Set the ticks on each axis
ax.set_xticks(np.arange(50, 600, 50))

## Display the plot
plt.show()
```

```python
## Rugplot Plotting with configure of layout (use text to set the plot title)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

## Input the data into pandas DataFram
tipDF = pd.read_excel("tips.xlsx")

## Define plot layout
fig, ax = plt.subplots(figsize=(10, 3))

## Make the rugplot
sns.rugplot(ax = ax, a=tipDF['total_bill'], height=0.25, lw=0.5, c = sns.xkcd_rgb["orange"])

## Configure the Rugplot outlook
ax.set_yticks([])      # Remove lable of y-axis
ax.set_xlim(0, 600)
ax.set_xlabel('Total Bills (\$)', fontsize=14)
ax.text(300, 0.45, 'Rug Plot on Total Bills Distribution', ha='center', fontsize=14, \
        color=sns.xkcd_rgb["pale red"])  # specify the title

# Clean up plot appearance
sns.despine(ax=ax, left=True, offset=2, trim=True)

## Display the plot
plt.show()
```

-------------
<ul style="color:SlateBlue;font-size=14px;">
    <li>We may also be able to use RugPlot to compare and contrast tow data directly by align the plots horizontally or vertically</li>
    <li>To display two RugPlot on the same figure, we have to specify our wish on the plotting code slightly when created the fig and ax with the subplots function as</li>
    <span style="color:white;">......</span><span style="color:orange;">fig, ax = plt.subplots (nrows=2, ncols=1, sharex=True)</span><br><br>
    <li>Options that we can be considered:</li>
    <ul style="color:DodgerBlur;font-size:14px;">
        <li><b>nrows</b>: the # of rows that will contain sub-plots.</li>
        <li><b>ncols</b>: the # of columns that will contain sub-plots.</li>
        <li><b>sharex</b>: whether the plots share the x-axis.</li>
        <li><b>sharey</b>: whether the plots share the y-axis.</li>
    </ul>
    </ul>
    
-----------------    

```python
## Rugplot Plotting for more than one plots
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

## Input the data into pandas DataFram
tipDF = pd.read_excel("tips.xlsx")

# Collect our two data sets as NumPy arrays
dTips = []
dTips.append(tipDF.loc[tipDF.time == 'Lunch', ['total_bill']])
dTips.append(tipDF.loc[tipDF.time == 'Dinner', ['total_bill']])
                      
## Define the plot layout
fig, ax = plt.subplots(figsize=(10, 4.0), nrows=2, ncols=1, sharex=True)

## Define plot dataset, colors and titles
clr = [sns.xkcd_rgb["pale red"], sns.xkcd_rgb["denim blue"]]
ttl = ['Lunch Time', 'Dinner Time']

## Make individual plot configurations
for idx, data in enumerate(dTips):
    ## Rugplot, with set rug height, thickness, and color
    sns.rugplot(ax = ax[idx], a=data, height=0.25, lw=0.5, c = clr[idx])  # specify height, thickness, and color
    ax[idx].set_yticks([])                                                # remove the y-axis
    ax[idx].set_xlim(0, 600)                                              # specify the x-axis limits and label
    ax[idx].set_xlabel('Total Bill (\$)', fontsize=14)
    ax[idx].text(300, 0.45, ttl[idx], ha='center', fontsize=14, color=clr[idx])   # title of the subplot             

    # Clean up each plot
    sns.despine(ax=ax[idx], left=True, offset=2, trim=True)
    
## Display the plots
plt.show()
```

<h3 style="color:blue;"><b>Box Plots</b></h3>
<ul style="color:SlateBlue;font-size:14px;">
    <li>Itprovides a graphical description of the distribution of data.</li>
    <li>The box plot is the box that encloses the middle 50% of the data (one edge is drawn at the 25<sup>th</sup> percentile, and the opposite edge is drawn at the 75<sup>th</sup> percentile), with a line (and an optional indentation mark) drawn at the median of the distribution.</li>
    <li>The box sometimes hasoutliners ()extending from either end that are used to show the extremes of the data distribution.</li>
    <li>To make boxplots, we will use the Seabornboxplot function, which can extract a column from a Pandas DataFrame to display the data.</li>
    </ul>
    
--------------------


```python
## Boxplot to display the distributions of total bills
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

## Input the data into pandas DataFram
tipDF = pd.read_excel("tips.xlsx")

# Define boxplot layout
fig, ax = plt.subplots(figsize=(10, 2.5))

# sns.set_style("whitegrid")
ax = sns.boxplot(x=tipDF["total_bill"], notch=True)

# Set x-axis length, label, and title the plot
ax.set_xlim(0, 600)
ax.set_xlabel('Total Bill (\$)', fontsize=14)
ax.set_title('Distribution of Total Bills', ha='center', fontsize=14)

## clean up each plot and show the final version of plot
sns.despine(left=True, offset=2, trim=True)
plt.show()
```

<h3 style="color:blue;"><b>Swarm Plots</b></h3>
<ul style="color:SlateBlue;font-size:14px;">
    <li>Seaborn library also provides support for visualizing a univariate distribution as a function of a second categorical variable.</li>
    <li>This visualization, called a swarm plot, displays the univariate distribution broken out by the categorical variable, but adds jitter (or random perturbation) to the distribution so that the full range of data is viewable.</li>
    </ul>
    
-------------------

```python
## Swarm plot to compare distributions of total bills on days
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

## Input the data into pandas DataFram
tipDF = pd.read_excel("tips.xlsx")

## Define the plot layout
fig, ax = plt.subplots(figsize=(8, 8))

## Configure the outlook of the Swarmplot
sns.swarmplot(ax=ax, x='day', y='total_bill', data=tipDF)

## Define x-axis limits and labels
ax.set_ylim(0, 600)
ax.set_xlabel('Day of the Week', fontsize=14)
ax.set_ylabel('Total Bill (\$)', fontsize=14)
ax.set_title('Total Bill Comparison by Days', ha='center', fontsize=16)

## Clean up each plot and display the plot
sns.despine(ax=ax, offset=10, trim=True)
plt.show()
```

```python
## Display Total Distributions with BoxPlot and RugPlot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

## Input the data into pandas DataFram
tipDF = pd.read_excel("tips.xlsx")

# Define boxplot layout
fig, ax = plt.subplots(figsize=(10, 2.5))

# Place both boxplot and rugplot on the same figure
ax = sns.boxplot(x=tipDF["total_bill"], color='#95d0fc', notch=True)    # you may enter the color code yourself
ax = sns.rugplot(a=tipDF['total_bill'], height=2, lw=0.5, \
            color = sns.xkcd_rgb["pale red"], alpha=0.5)

# Set x-axis length, label, and title the plot
ax.set_xlim(0, 600)
ax.set_xlabel('Total Bill (\$)', fontsize=14)
ax.set_title('Distribution of Total Bills', ha='center', fontsize=14)

## clean up each plot and show the final version of plot
sns.despine(left=True, offset=2, trim=True)
plt.show()
```
