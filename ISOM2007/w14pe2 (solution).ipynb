```python
## Display Total Distributions with BoxPlot and Swarm Plot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

## Input the data into pandas DataFram
tp = pd.read_csv("tips.csv")

## Define the plpt layout
fig, ax = plt.subplots(figsize=(6, 6))

## Configure the layout of the plots
sns.swarmplot(ax=ax, x='sex', y='tip', data=tp)
sns.boxplot(ax=ax, x='sex', y='tip', data=tp, notch=True, color=sns.xkcd_rgb["tiffany blue"])

## Define x-axis limits and labels
ax.set_ylim(0, 110)
ax.set_yticks(np.arange(0,120, 10))
ax.set_xlabel('Day of the Week', fontsize=12)
ax.set_ylabel('Total tips (\$)', fontsize=12)
ax.set_title('Comparison of Tips Distribution by Sex', ha='center', fontsize=14, color=sns.xkcd_rgb["royal blue"])

## Clean up each plot and display the plot
sns.despine(ax=ax, offset=10, trim=True)
plt.show()
```
