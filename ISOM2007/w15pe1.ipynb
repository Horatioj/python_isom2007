<h3 style="color:blue;"><b>Scatter Plots</b></h3>
<ul style="color:SlateBlue;">
    <li>Scatter plots, one of the most important visualization skills for a data scientist.</li>
    <li>It is a technique for visually exploring two-dimensional data. 
It enables us to determine whether the two dimensions of a dataset share a positive, negative, or null correlation as well as aids in the identification of outliers.</li>
    <li>To make a scatter plot in Python, we can either use the plot() method as discussed in the last module, or we can use the scatter() method, while scatter() method provides greater flexibility when making scatter plots.</li>
    </li>Recall from the previous lecture we will first need to set up the working environment of the Notebook :</li>
    </ul>
    
-------------------------

```python
# Set up the Notebook working environment
%matplotlib inline

## Standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

## We do this to ignore several specific Pandas warnings
import warnings
warnings.filterwarnings("ignore")

## Use default white plot style
sns.set(style="white")
```

---------
<h3 style="color:blue;"><b>Positive Correlation</b></h3>
<ul style="color:SlateBlue;font-size:14px;">
    <li>Scatter plot where the vertical values increase as the horizontal value increases.</li>
    </ul>
    
--------------

```python
### Standard imports been hidden to save space ####
## Now we create our figure and axes for the plot we will make.
fig, ax = plt.subplots(figsize=(8,6))

## Now we generate something to plot. 
x = np.linspace(0,100)
y = x + np.random.uniform(-10, 10, 50)

## Specify the type of plot
ax.scatter(x, y) 

# Set our axis labels, specify the limit, and ticks
ax.set_xlabel("X Axis", fontsize=14)
ax.set_ylabel("Y Axis", fontsize=14)
ax.set_xlim(-20, 120)
ax.set_ylim(-20, 120)
ax.set_xticks(np.arange(0, 120, 20))
ax.set_yticks(np.arange(0, 120, 20))
ax.set_title("A positive correlation scatter plot!", fontsize=14)

## Clean up final result and display the plot inline
sns.despine(offset=2, trim=True)
plt.show()
```

---------
<h3 style="color:blue;"><b>Negative Correlation</b></h3>
<ul style="color:SlateBlue;font-size:14px;">
    <li>Scatter plot where the vertical values increase as the horizontal value decreases or vice versa.</li>
    </ul>
    
--------------

```python
### Standard imports been hidden to save space ####
## Now we create our figure and axes for the plot we will make.
fig, ax = plt.subplots()

## Now we generate something to plot. 
x = np.linspace(0,100)
y = 100 - x + np.random.uniform(-10, 10, 50)

## Specify the type of plot
ax.scatter(x, y) 

# Set our axis labels, specify the limit, and ticks
ax.set_xlabel("X Axis", fontsize=14)
ax.set_ylabel("Y Axis", fontsize=14)
ax.set_xlim(-20, 120)
ax.set_ylim(-20, 120)
ax.set_xticks(np.arange(0, 120, 20))
ax.set_yticks(np.arange(0, 120, 20))
ax.set_title("A negative correlation scatter plot!", fontsize=14)

## Clean up final result and display the plot inline
sns.despine(offset=2, trim=True)
plt.show()
```

```python
## Standard imports been hidden to save space ####
## Scatter plot to display the correlation between mid-term with exam scores
## Input the Data set, and define our two dimensions
gest = pd.read_excel("GEST0000.xlsx")
mdf = gest['Mid-term']
edf = gest['Exam']

## Create our figure and axes for the plot and plot the data
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(mdf, edf)

# Set our axis labels and title
ax.set_xlabel("Mid-term Scores", fontsize=12)
ax.set_ylabel("Exam Scores", fontsize=12)
ax.set_title("Comparison of Mid-term with Exam Scores!", fontsize=14)
ax.set_xlim(0, 110)
ax.set_ylim(0, 110)

sns.despine(offset=10, trim=True)
plt.show()
```

---------------
<ul style="color:SlateBlue;">
    <li>Scatter plot to compare the correlation between Mid-term/Exam scores with Overall scores.</li>
    </ul>
    
-----------------

```python
## Standard imports been hidden to save space ####
## Scatter plot to display the correlation between Mid-term/Exam with Overall scores
## Input the Data set, and define our two dimensions
gest = pd.read_excel("GEST0000.xlsx")
mdf = gest['Mid-term']
edf = gest['Exam']
ovl = gest['Overall']
x = np.arange(0,100, 0.5)

## Create our figure and axes for the plot and plot the data
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(ovl, mdf, color=sns.xkcd_rgb["pale red"], marker='p', label='Mid-term')
ax.scatter(ovl, edf, color=sns.xkcd_rgb["slate blue"], marker='d', label='Exam')
ax.plot(x, x, color="green", ls="dashed", label="Relation")

# Set our axis labels and title
ax.set_xlabel("Overall Scores", fontsize=12)
ax.set_ylabel("Mid-term/Exam Scores", fontsize=12)
ax.set_title("Comparison of Mid-term/Exam and Overall Scores!", fontsize=14)
ax.set_xlim(0, 110)
ax.set_ylim(0, 110)
ax.legend(loc='upper left', fontsize=14)

sns.despine(offset=20, trim=True)
plt.show()
```

<h3 style="color:blue;"><b>Reg Plots</b></h3>
<ul style="color:SlateBlue;font-size:14px;">
    <li>regplot from the Seaborn module performs a simple and  quickly display of a scatter plot between the x and y features and the indication of the good fit.</li>
    <br>
    <li>Example of regplot with fit_reg=False</li>
    </ul>
    
--------------

```python
## Standard imports been hidden to save space ####
## Scatter plot to display the correlation between mid-term with exam scores
## Input the Data set, and define our two dimensions
gest = pd.read_excel("GEST0000.xlsx")

## Create our figure and axes for the plot and plot the data
fig, ax = plt.subplots(figsize=(5, 5))

## ax.scatter(mdf, edf)
ax = sns.regplot(x='Mid-term', y='Exam', data=gest, fit_reg=False)  # instead use of scatter plot

# Set our axis labels and title
ax.set_xlabel("Mid-term Scores", fontsize=12)
ax.set_ylabel("Exam Scores", fontsize=12)
ax.set_title("Comparison of Mid-term with Exam Scores!", fontsize=14)

sns.despine(offset=10, trim=True)
plt.show()
```

----------
<ul style="color:SlateBlue;font-size:14px;">
    <li>Example of regplot with fit_reg=True & ci=95</li>
    </ul>
    
-------------------

```python
## Standard imports been hidden to save space ####
## Scatter plot to display the correlation between mid-term with exam scores
## Input the Data set, and define our two dimensions
gest = pd.read_excel("GEST0000.xlsx")

## Create our figure and axes for the plot and plot the data
fig, ax = plt.subplots(figsize=(5, 5))

## ax.scatter(mdf, edf)
ax = sns.regplot(x='Mid-term', y='Exam', data=gest, fit_reg=True, ci=95)  # instead use of scatter plot

# Set our axis labels and title
ax.set_xlabel("Mid-term Scores", fontsize=12)
ax.set_ylabel("Exam Scores", fontsize=12)
ax.set_title("Comparison of Mid-term with Exam Scores!", fontsize=14)

sns.despine(offset=10, trim=True)
plt.show()
```

----------
<ul style="color:SlateBlue;">
    <li>Determine the Best Fit Regression Line, y = ax + b, to verify the plot.</li>
    </ul>
    
------------

```python
# Perform simple linear regression
from scipy import stats as sts

gest = pd.read_excel("GEST0000.xlsx")
mdf = gest['Mid-term']
edf = gest['Exam']

# Compute our slope, intercept, and r value
slope, intercept, r_value, p_value, slope_std_error = sts.linregress(mdf, edf)

# Display results
print(f'\nBest fit line: y = {slope:4.2f} x + {intercept:5.3f}')
print(f'Pearsonr correlation = {r_value:5.3f}')
```
