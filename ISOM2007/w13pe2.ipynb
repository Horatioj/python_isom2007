```python
## Create a Pivot Table and generate a pie chart using the day column 
## as the vertical index and the sex column as the horizontal index
import pandas as pd

## Read data from Excel file
tipsDF = pd.read_excel('tips.xlsx')

## Generate the pivot table accordingly
tb = tipsDF.pivot_table('total_bill', index='day', columns='sex')

## Plot a pie chart with the data from the pivot table
tb.plot.pie(subplots=True, figsize=(11,6))
```

<h3 style="color:blue;"><b>Matplotlib Basics</b></h3>
<ul style="color:SlateBlue;">
    <li>The discussions in the coming 3 ~ 4 lectures, we will dig into the matplotlib module, which forms the foundation for almost all Python plotting, to understand how all aspects of a simple visualization are created and modified.</li>
    <li>The matplotlib plotting library provides powerful data visualization functionality.</li>
    <li>The recommended method to use matplotlib is via the pyplot interface as it provides an object-oriented access to the full matplotlib library.</li><br>
    <li>Steps of Using matplotlib to Plot a Chart</li>
    <ul style="color:DodgerBlue;">
        <li>Import the pyplot interface</li>
        <span style="color:white;">....</span><span style="color:orange;">import matplotlib.pyplot as plt</span><br><br>
 <li>Create Figure object that will allow us to control the global appearance of our visualization</li>
<span style="color:white;">....</span><span style="color:orange;">fig = plt.figure()</span><br><br>
        <li>Create an Axes object that allows us to make an actual plot within our figure</li>
<span style="color:white;">....</span><span style="color:orange;">ax = fig.add_subplot()</span><br><br>
 <li>We may merge step 2 & 3 into one as:</li>
<span style="color:white;">....</span><span style="color:orange;">fig, ax = plt.subplots()</span><br>
        
----------------------------

```python
import matplotlib.pyplot as plt
import numpy as np

## Now we create our figure and axes for the plot we will make.
fig, ax = plt.subplots(figsize=(10,4))

## Plot a straight according to linear equation y = 6 + 2x
x = np.linspace(0,10)
y = 6 + 2 * x

## Label the axis and title
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_title("Graph of Linear Function")

## Plot according to the data generated, connecting the points with lines
ax.plot(x, y)

## We call plt.show() to show the chart (optional, but good to have)
plt.show()
```

<h3 style="color:blue;"><b>Restricting the Axes Range</b></h3>
<ul style="color:SlateBlue;">
    <li>Matplotlib allows us the flexibility to control over the outlook of the plot.</li>
    <li>Two other items that we may want to specify in our plots are:</li>
    <ol style="color:DodgerBlue;">
        <li>Degree to which our plot fills the subplot</li>
        <ul style="color:SkyBlue;">
            <li>This can be done by changed the range displayed in either the x-axis, the y-axis, or both by setting the limits to matplotlib to change the span as specified:</li>
        </ul>
        <span style="color:white;">..............</span><span style="color:orange;">ax.set_xlim(-2, 12)</span><br>
 <span style="color:white;">..............</span><span style="color:orange;">ax.set_ylim(-10, 40)</span><br><br>
        <li>How the numbers are displayed on the axes</li>
        <ul style="color:SkyBlue;">
            <li>This can be done by specify intervals on the axes, which are known as ticks, we can control each axis independently by using the set method to specify the x-axis labels include 0, 5, 10, and 15, while y-axes labels from -15 to 10 with increments of 5.</li>
        </ul>
 <span style="color:white;">..............</span><span style="color:orange;">ax.set_xticks(np.arange(0, 15, 5)</span><br>
 <span style="color:white;">..............</span><span style="color:orange;">ax.set_yticks(np.arange(-10, 40, 5)</span><br>
        </ul>

```python
import matplotlib.pyplot as plt
import numpy as np

## Now we create our figure and axes for the plot we will make.
fig, ax = plt.subplots(figsize=(10,4))

## Plot a straight according to linear equation y = 6 + 2x
x = np.linspace(0,10)
y = 6 + 2 * x

## Label the axis and title
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_title("Graph of Linear Function")

# Change the axis limits displayed in our plot
ax.set_xlim(-2, 12)
ax.set_ylim(-10, 40)

# Change the ticks on each axis and the corresponding numerical values that are displayed
ax.set_xticks(np.arange(0, 15, 5))
ax.set_yticks(np.arange(-15, 10, 5))

## Plot according to the data generated, connecting the points with lines
ax.plot(x, y)

## We call plt.show() to show the chart (optional, but good to have)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

fig, ax = plt.subplots(figsize=(20, 8))

ax.set_xlabel("Costs of Advertisement", fontsize=20)
ax.set_ylabel("Marginal Profits", fontsize=20)
ax.set_title("Sensitivity Analysis", fontsize=24)
ax.set_xlim(-5, 2)
ax.set_ylim(-120, 40)
ax.set_xticks(np.arange(-4, 3, 0.5))
ax.set_yticks(np.arange(-120, 2, 10))
x1 = np.linspace(-4, 2)
y1 = 18*x1 - 5

x2 = np.linspace(-4, 2)
y2 = -7 * x2 ** 2 + 3

ax.plot(x1, y1)
ax.plot(x2, y2)

plt.show()

plt.savefig("my_plot.svg")
```
