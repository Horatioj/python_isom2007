```python
## Input the data file tips.xlsx in Pandas DataFrame and plot
## the histogram on total_bill and tip
import pandas as pd
import matplotlib.pyplot as plt

## Input the data file tips.xlsx with Pandas read_excel()
tipDF = pd.read_excel("tips.xlsx")

## Overplot histograms of the total_bill and tips columns.
tipDF[['total_bill', 'tip']].plot.hist(bins=10, alpha=0.5, figsize=(12, 6)) 

## Display the plot
plt.show()
```

<p style="color:SlateBlue;font-size:14px;">In the empty Code cell below, write and execute code to make a plot for two different quadratic functions according to the followings:</p>
<ol style="color:DodgerBlue;font-size:14px;">
    <li>y<sub>1</sub> = 4x<sup>2</sup> + 12x - 25.</li>
    <li>y<sub>2</sub> = -3x<sup>2</sup> - 13x + 37.</li>
    <li>Give the plot a suitable title.</li>
    <li>Add a legend, place it at a place not blocking the curves.</li>
    <li>Label the x and y axis to be Height (cm) and Mass (kg), respectively.</li>
    <li>Specify the x limit and y limit to allow some spaces at the 4 edges.</li>
    <li>Offset the axises with some spaces to enhance the readability</li>
    <li>Set the color for y<sub>1</sub> to brown and y<sub>2</sub> to orange</li>
    </ol>
    
-------------------------

```python
## Plot of two different cubic functions
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns            

## Define some color for the graph
g_colors = [sns.xkcd_rgb['orange'], sns.xkcd_rgb['brown']]

## Seaborn specifications
sns.set_context("poster", font_scale=0.5)      # specify the context with font size
#sns.set_style("white")
#sns.set_style("ticks")

fig, ax = plt.subplots(figsize=(10,5))
x = np.linspace(-8, 6)
y1 = 4 * x**2 + 12 * x - 25
y2 = -3 * x**2 - 13 * x + 37
ax.plot(x, y1, label = 'Plot 1', ls=':', color=g_colors[1])           # specify the style & color
ax.plot(x, y2, label = 'Plot 2', ls="dashdot", color=g_colors[0])
        
## Set labels of axis
ax.set_xlabel("Height (cm)")
ax.set_ylabel("Mass (kg)")

## Set limit of the range on plot
ax.set_xlim(-10, 8)
ax.set_ylim(-120, 200)

## Set the ticks on each axis
ax.set_xticks(np.arange(-8, 8, 2))
ax.set_yticks(np.arange(-100, 250, 50))

plt.legend(loc='lower center')             # include the legend and specify its location
    
## Set the title
ax.set_title("Plot of Height vs Mass")

## Seaborn specific modifications
sns.despine(offset=20, trim=False)

plt.show()
```
