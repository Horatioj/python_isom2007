```python
##1.


## input your purchasing price and selling price and put two prices into a list. Return the list.
def input_():
    pur_price = eval(input('plese input purchase price: '))
    sell_price = eval(input('please input selling price: '))
    list_price = [pur_price, sell_price]
    return list_price

## calculate the gross price and return gross profit
def gross(pur_price, sell_price):
    gross = sell_price - pur_price
    return (gross)

## calculate the percentage gross profit and display 
def gross_profit(pur_price, gross):
    percent_gross = gross/pur_price * 100
    print('the percentage gross profit is %.2f%%' % percent_gross)
    
## calculate the profit margin and display
def profit_margin(gross, sell_price):
    mar_profit = gross/sell_price * 100
    print('the margin profit is %.2f%%' % mar_profit)
```

```python
##3
'''For question 3, in my views, it is hard to halve the second price simply because 
it is input at the second place unless it is input strictly in order. Additionally, since the question 
requires us to compare between the prices as well. Besides, regarding to this question, 
it sounds like a little strange to halve the even-th number of goods you buy, why not just place 
the expensive one on the even-th order? Also it is strange to just consider buying 2 goods.'''


## import math to utilize the method of '.floor()'
import math

def q3Main():
    while True:
        list_input = Input_()
        Tuple_Compare = Compare(list_input[0], list_input[1])
        Display_(Tuple_Compare[0], Tuple_Compare[1])
        Amt_after = amtDiscount(Tuple_Compare[0], Tuple_Compare[1])
        amtTax(Amt_after)
        determing = input("Would like to keep inputing prices(no for quit)?(yes/no)")
        if determing == 'no':
            break

## input the price of your suits
def Input_():
    suit1 = eval(input("please input the price of your first suit: "))
    suit2 = eval(input("please input the price of your second suit: "))
    list_input = [suit1, suit2]
    return list_input

## compare the price of the two suits, and make suit1 larger than suit2
def Compare(suit1, suit2):
    if suit1 >= suit2:
        return (suit1, suit2)
    else:
        suit1, suit2 = suit2, suit1
        return (suit1, suit2)

## output relevant information, suit1 is more expansive than suit2
def Display_(suit1, suit2):
    print("Invoice of purchase")
    print("{0:<20s} {1:^3}{2:>10.2f}".format('Suit1', ':', suit1))
    print("{0:<20s} {1:^3}{2:>10.2f}".format('Suit2', ':', suit2))
    print("{0:<20s} {1:^3}{2:>10.2f}".format('Amount of purchase', ':', (suit1 + suit2)))

## calculate and display the discount amount
def amtDiscount(suit1, suit2):
    discount = 0.5 * suit2
    Amt_after = suit1 + suit2 - discount
    print("{0:<20s} {1:^3}{2:>10.2f}".format('Discount', ':', discount))
    print("{0:<20s} {1:^3}{2:>10.2f}".format('After discount', ':', Amt_after))
    return Amt_after

## calculate and display the tax on purchasing after discount and final cost charged
def amtTax(Amt_after):
    Amt_tax = Amt_after * 0.07
    Amt_charge = Amt_after * 1.07
    print("{0:<20s} {1:^3}{2:>10.2f}".format('Amount of tax', ':', (math.floor(Amt_tax))))
    print("{0:<20s} {1:^3}{2:>10.2f}".format('Amount charged', ':', (math.floor(Amt_charge))))

## execution
q3Main()
```

```python
##4


## main function to call following functions
def q4Main():
    list_s = Input_()
    scores = Calc_score(list_s)
    Display_s(list_s)
    Grade_(scores)
    
## input student's performance in accordance with the question, bb912345 56 78 69 80
## creat a list
def Input_():
    sInput = input("please enter the student's ID and scores: ")
    list_s = sInput.split()
    return list_s

## calculate final scores
def Calc_score(list_s):
    list_s = list_s[1:]
    list_s = list(map(int, list_s))
    scores = list_s[0] * 0.1 + list_s[1] * 0.2 + list_s[2] * 0.3 + list_s[3] * 0.4
    return scores

## display list elements
def Display_s(list_s):
    print("Student ID: %s" % list_s[0])
    print("In-class excercises: %s" % list_s[1])
    print("Assignments: %s" % list_s[2])
    print("Mid-term: %s" % list_s[3])
    print("Exam: %s" % list_s[4])
    print("Final scores: %s" % Calc_score(list_s))
    
## find the final grades according to your scores
def Grade_(scores):
    if scores >= 90:
        grade = 'A'
    elif scores >= 80:
        grade = 'B'
    elif scores >= 70:
        grade = 'C'
    elif scores >= 60:
        grade = 'D'
    elif scores >= 50:
        grade = 'E'
    else: 
        grade = 'F'
    print("overall grade: " + grade)
 ## execuate functions
q4Main()
```
