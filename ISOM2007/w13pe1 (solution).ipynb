<p style="color:SlateBlue;">In the empty Code cell below, write and execute code to make a four by five matrix that contains the integers 2, 4, 6, ..., 40. Slice and display:</p>
<ol style="color:DodgerBlue;">
    <li>the third row,</li>
    <li>the fourth column, and</li>
    <li>the element at the position (4, 4).</li>
    </ol>
    
-------------------------

```python
import numpy as np

## Create a 4x5 Matrix contains the integers 2, 4, 6, ...., 40
aData = np.arange(2, 41, 2).reshape(4, 5)
print("Initial Matrix:\n{}".format(aData))

## Slice and display the third row
print("\nThe 3rd row of the matrix:\n{}".format(aData[2]))

## Slice and display the fourth column
print("\nThe 4th column of the matrix:\n{}".format(aData[:, 3]))

## Slice and display the element at the position (4, 4).
print("\nThe element at (4,4) of the matrix: {}".format(aData[3,3]))
```

<p style="color:SlateBlue;">In the empty Code cells below, write and execute code to make a four by four matrix that contains the integers 1 through 16, then perform the following:</p>
<ol style="color:DodgerBlue;">
    <li>Add five to the first row, ten to the second row, fifteen to the third row, twenty to the fourth row, and display the resulting matrix. </li>
    <li>Next apply the cos function to the odd rows, and the sin function to the even rows, and display the resulting matrix.</li>
    <li>Finally, compute and display the cumulative sum for each row, and the cumulative product for each column.</li>
    </ol>

----------------------------------

```python
## Add five to the first row, ten to the second row, fifteen to the third row, 
## twenty to the fourth row, and display the resulting matrix.
import numpy as np
import numpy.ma as ma

## Create the 4x4 matrix that contains 1 to 16
aData = np.arange(1, 17).reshape(4,4)
print("Initial Matrix:\n{}".format(aData))

## Add 5 to the 1st row
mask = ma.asarray(aData, dtype=bool)    # you may just create another matrix of the same size
mask[:,:] = False     # set all entries to False
mask[0] = True        # set the entries on 1st row to True
aData[mask] += 5
print("\nMatrix after 1st row +5 :\n{}".format(aData))

## Add 10 to the 2nd row
mask[:,:] = False
mask[1] = True
aData[mask] += 10
print("\nMatrix after 2nd row +10 :\n{}".format(aData))

## Add 15 to the 3rd row
mask[:,:] = False
mask[2] = True
aData[mask] += 15
print("\nMatrix after 3rd row +15 :\n{}".format(aData))

## Add 20 to the 4th row
mask[:,:] = False
mask[3] = True
aData[mask] += 20
print("\nMatrix after 4th row +20 :\n{}".format(aData))

## Appy cos function to odd rows
aData = np.arange(1,17).reshape(4,4).astype(float)   # have the original matrix in float
print("\n{0}-\n\nInitial Matrix:\n{1}".format(15*"- * ", aData))
mask = (aData > 999999) # alternative way to mask be false, assume 999999 > all values in aData
mask[0] = True          # set odd row to True
mask[2] = True          # set odd row to True
aData[mask] = np.cos(aData[mask])
print("\nMatrix after apply sin on row 1 & 3 :\n{}".format(aData))

## Appy sin function to even rows
mask[:,:] = False
mask[1] = True
mask[3] = True
aData[mask] = np.sin(aData[mask])
print("\nMatrix after appy cos on row 2 & 4 :\n{}".format(aData))

## compute and display the cumulative sum for each row, and
## the cumulative product for each column.
aData = np.arange(1,17).reshape(4,4)          # have the original matrix
print("\n{0}-\n\nInitial Matrix:\n{1}".format(15*"- * ", aData))

# cummulative sum on each row over the columns
print("\nThe cumulative sum for each row: \n{}".format(np.cumsum(aData, axis=1)))

# cummulative product on each column over the rows
print("\nThe cumulative product for each column: \n{}".format(np.cumprod(aData, axis=0)))
```

```python
## Alternative Solution - provided that there exist consistent pattern on the entries ##

import numpy as np

## Create two matrix of the same size
control = np.arange(1, 17).reshape(4,4)
aData = np.arange(1, 17).reshape(4,4)
print("Initial Matrix:\n{}".format(aData))

## Add 5 to 1st row, 10 to 2nd, 15 to 3rd, and 20 to 4th row
mask = (control <= 4)
aData[mask] += 5
print("\nMatrix after 1st row +5 :\n{}".format(aData))

mask = np.logical_and(control >= 5, control <= 8)
aData[mask] += 10
print("\nMatrix after 2nd row +10 :\n{}".format(aData))

mask = np.logical_and(control >= 9, control <= 12)
aData[mask] += 15
print("\nMatrix after 3rd row +15 :\n{}".format(aData))

mask = (control >= 13)
aData[mask] += 20
print("\nMatrix after 4th row +20 :\n{}".format(aData))

## Appy cos function to odd rows and sin function to even rows
aData = np.arange(1,17).reshape(4,4).astype(float)
print("\n{0}-\n\nInitial Matrix:\n{1}".format(15*"- * ", aData))
mask1 = (control <= 4)
mask2 = np.logical_and(control >= 9, control <= 12)
mask = np.logical_or(mask1, mask2)
aData[mask] = np.cos(aData[mask])
print("\nMatrix after apply cos on row 1 & 3 :\n{}".format(aData))

mask1 = np.logical_and(control >= 5, control <= 8)
mask2 = (control >= 13)
mask = np.logical_or(mask1, mask2)
aData[mask] = np.sin(aData[mask])
print("\nMatrix after apply sin on row 2 & 4 :\n{}".format(aData))
```
